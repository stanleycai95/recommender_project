```python
import pandas as pd

import numpy as np
```

```python
img_links = pd.read_csv('listings_img_quality_3_2023_03_01.csv', header=None)
```

```python
img_links.columns = ['img_link', 'is_sold']

img_links = img_links.sample(frac=1).reset_index(drop=True)
```

```python
img_links_sampled = pd.concat([img_links.loc[img_links.is_sold == 1], img_links.loc[img_links.is_sold == 0].sample(frac=0.027)])

#img_links_sampled = img_links.copy()
```

```python
img_links_sampled['is_sold'].mean()
```

```python
img_links_sampled.reset_index(drop=True, inplace=True)
```

```python
img_links_sampled['folder_link'] = None
```

```python
img_links_sampled = img_links_sampled.sample(frac=1).reset_index(drop=True)
```

```python
from PIL import Image
import requests
```

```python
from PIL import Image
import requests

for i in range(1000):
  print(i)
  try:
    url = img_links_sampled.iloc[i]['img_link']
    im = Image.open(requests.get(url, stream=True).raw).convert('RGB')
    if img_links_sampled.iloc[i]['is_sold'] == 1:
      im.save(f"./SOLD_{i}.jpg")
      img_links_sampled.at['folder_link', i] = f"./SOLD_{i}.jpg"
    elif img_links_sampled.iloc[i]['is_sold'] == 0:
      im.save(f"./FAILED_{i}.jpg")
      img_links_sampled.at['folder_link', i] = f"./FAILED_{i}.jpg"
  except:
    print("Error")
```

```python
from fastai.vision.all import *
from fastai.text.all import *
from fastai.collab import *
from fastai.tabular.all import *
```

```python
import os

paths = os.listdir('./')

pd.Series([1 if 'SOLD' in path else 0 for path in paths]).mean()
```

```python
path = Path('./')
```

```python
def is_car_sold(x): return 'SOLD' in x
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_car_sold, item_tfms=Resize(224))

f1score = F1Score()
roc_auc = RocAucBinary()
learn = vision_learner(dls, vgg19_bn, metrics=[error_rate, roc_auc, f1score])
learn.fine_tune(2, freeze_epochs=5)
```

```python
img = PILImage.create('./SOLD_5.jpg')
img
```

```python
is_sold, _, probs = learn.predict(img)
print(f"Is this vehicle sold?: {is_sold}.")
print(f"Probability it's sold: {probs[1].item():.6f}")
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```
